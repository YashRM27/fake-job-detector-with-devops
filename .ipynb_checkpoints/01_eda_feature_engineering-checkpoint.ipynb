# Fake Job Posting Detector
## Notebook 1: EDA & Feature Engineering

**Dataset:** EMSCAD – Real or Fake Job Postings (Kaggle)  
**Target:** `fraudulent` — 0 = Real, 1 = Fake  
**Goal:** Understand the data, handle imbalance, engineer features for modelling


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False

print('Libraries loaded successfully')
```

## 1. Load Data

```python
# Place the downloaded CSV in data/ folder
df = pd.read_csv('../data/fake_job_postings.csv')

print(f'Shape: {df.shape}')
print(f'Columns: {list(df.columns)}')
df.head(3)
```

## 2. Basic Info & Dtypes

```python
df.info()
```

## 3. Target Distribution — Class Imbalance

```python
counts = df['fraudulent'].value_counts()
pct = df['fraudulent'].value_counts(normalize=True) * 100

print('Class distribution:')
print(f'  Real (0): {counts[0]:,}  ({pct[0]:.1f}%)')
print(f'  Fake (1): {counts[1]:,}  ({pct[1]:.1f}%)')
print(f'  Imbalance ratio: {counts[0]/counts[1]:.1f}:1')

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].bar(['Real', 'Fake'], [counts[0], counts[1]], color=['#2ecc71', '#e74c3c'], width=0.5)
axes[0].set_title('Job Posting Count by Class')
axes[0].set_ylabel('Count')
for i, v in enumerate([counts[0], counts[1]]):
    axes[0].text(i, v + 50, f'{v:,}', ha='center', fontweight='bold')

axes[1].pie([pct[0], pct[1]], labels=['Real', 'Fake'],
            colors=['#2ecc71', '#e74c3c'], autopct='%1.1f%%', startangle=90)
axes[1].set_title('Class Proportion')

plt.tight_layout()
plt.savefig('../notebooks/class_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 4. Missing Values Analysis

```python
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df) * 100).round(2)
missing_df = pd.DataFrame({'missing_count': missing, 'missing_pct': missing_pct})
missing_df = missing_df[missing_df['missing_count'] > 0].sort_values('missing_pct', ascending=False)

print(missing_df)

plt.figure(figsize=(10, 5))
plt.barh(missing_df.index, missing_df['missing_pct'], color='#e67e22')
plt.xlabel('Missing %')
plt.title('Missing Values by Column')
for i, v in enumerate(missing_df['missing_pct']):
    plt.text(v + 0.3, i, f'{v:.1f}%', va='center')
plt.tight_layout()
plt.show()
```

**Insight:** Missing salary_range, company_profile, and location are themselves signals — fake postings tend to leave these blank.

## 5. Missing Values by Class — Are Nulls Informative?

```python
# Check if missingness rate differs between real and fake postings
text_cols = ['company_profile', 'description', 'requirements', 'benefits', 'salary_range']

null_by_class = {}
for col in text_cols:
    if col in df.columns:
        null_by_class[col] = {
            'Real (0)': df[df['fraudulent']==0][col].isnull().mean() * 100,
            'Fake (1)': df[df['fraudulent']==1][col].isnull().mean() * 100
        }

null_class_df = pd.DataFrame(null_by_class).T
print('% Missing by class:')
print(null_class_df.round(1))

null_class_df.plot(kind='bar', color=['#2ecc71','#e74c3c'], figsize=(10,4))
plt.title('Missing % per Column by Class')
plt.ylabel('Missing %')
plt.xticks(rotation=30)
plt.legend(['Real', 'Fake'])
plt.tight_layout()
plt.show()
```

## 6. Categorical Feature Analysis

```python
cat_cols = ['employment_type', 'required_experience', 'required_education', 'industry', 'function']

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

for ax, col in zip(axes, ['employment_type', 'required_experience', 'required_education']):
    if col in df.columns:
        data = df.groupby([col, 'fraudulent']).size().unstack(fill_value=0)
        data.plot(kind='bar', ax=ax, color=['#2ecc71','#e74c3c'], legend=False)
        ax.set_title(col.replace('_', ' ').title())
        ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha='right', fontsize=8)

plt.suptitle('Categorical Features vs Fraud', y=1.02)
plt.tight_layout()
plt.show()
```

## 7. Binary Feature Analysis

```python
binary_cols = ['telecommuting', 'has_company_logo', 'has_questions']

fraud_rates = {}
for col in binary_cols:
    if col in df.columns:
        rate = df.groupby(col)['fraudulent'].mean() * 100
        fraud_rates[col] = rate
        print(f'{col}:')
        print(rate.round(2))
        print()
```

## 8. Text Length Analysis

```python
text_cols = ['title', 'description', 'company_profile', 'requirements', 'benefits']

for col in text_cols:
    if col in df.columns:
        df[f'{col}_len'] = df[col].fillna('').apply(len)

len_cols = [c for c in df.columns if c.endswith('_len')]

fig, axes = plt.subplots(1, len(len_cols), figsize=(18, 4))
for ax, col in zip(axes, len_cols):
    real = df[df['fraudulent']==0][col]
    fake = df[df['fraudulent']==1][col]
    ax.hist(real, bins=40, alpha=0.6, color='#2ecc71', label='Real', density=True)
    ax.hist(fake, bins=40, alpha=0.6, color='#e74c3c', label='Fake', density=True)
    ax.set_title(col.replace('_len','').replace('_',' ').title())
    ax.set_xlabel('Character count')
    ax.legend(fontsize=8)

plt.suptitle('Text Length Distribution: Real vs Fake')
plt.tight_layout()
plt.show()
```

## 9. Feature Engineering

```python
df_fe = df.copy()

# ── 9.1  Fill text nulls with empty string ──────────────────────────────────
text_cols = ['title', 'company_profile', 'description', 'requirements', 'benefits']
for col in text_cols:
    df_fe[col] = df_fe[col].fillna('')

# ── 9.2  Combine all text into one corpus field ──────────────────────────────
df_fe['text'] = (
    df_fe['title'] + ' ' +
    df_fe['company_profile'] + ' ' +
    df_fe['description'] + ' ' +
    df_fe['requirements'] + ' ' +
    df_fe['benefits']
)

# ── 9.3  Missingness flag features (nulls as signal) ────────────────────────
df_fe['missing_salary']          = df['salary_range'].isnull().astype(int)
df_fe['missing_company_profile'] = df['company_profile'].isnull().astype(int)
df_fe['missing_requirements']    = df['requirements'].isnull().astype(int)
df_fe['missing_benefits']        = df['benefits'].isnull().astype(int)
df_fe['missing_location']        = df['location'].isnull().astype(int)

# ── 9.4  Text length features ────────────────────────────────────────────────
df_fe['desc_len']    = df_fe['description'].apply(len)
df_fe['title_len']   = df_fe['title'].apply(len)
df_fe['text_len']    = df_fe['text'].apply(len)

# ── 9.5  Red flag keyword count ──────────────────────────────────────────────
RED_FLAGS = [
    'urgent', 'immediate', 'no experience', 'work from home',
    'earn money', 'guaranteed', 'wire transfer', 'send resume to',
    'investment required', 'personal details', 'multi-level',
    'unlimited earning', 'be your own boss', 'no degree required'
]

def count_red_flags(text):
    text_lower = text.lower()
    return sum(1 for flag in RED_FLAGS if flag in text_lower)

df_fe['red_flag_count'] = df_fe['text'].apply(count_red_flags)

print('Feature engineering complete.')
print(f'New columns added: text, missing_*, desc_len, title_len, text_len, red_flag_count')
print(f'\nRed flag count by class:')
print(df_fe.groupby('fraudulent')['red_flag_count'].describe().round(2))
```

## 10. Encode Categorical Features

```python
from sklearn.preprocessing import LabelEncoder

cat_cols_to_encode = ['employment_type', 'required_experience', 'required_education', 'industry', 'function']

le = LabelEncoder()
for col in cat_cols_to_encode:
    if col in df_fe.columns:
        df_fe[col] = df_fe[col].fillna('Unknown')
        df_fe[col + '_enc'] = le.fit_transform(df_fe[col])

print('Categorical encoding done.')
```

## 11. Define Final Feature Sets

```python
# Structured (non-text) features
STRUCTURED_FEATURES = [
    'telecommuting', 'has_company_logo', 'has_questions',
    'missing_salary', 'missing_company_profile', 'missing_requirements',
    'missing_benefits', 'missing_location',
    'desc_len', 'title_len', 'text_len', 'red_flag_count',
    'employment_type_enc', 'required_experience_enc',
    'required_education_enc', 'industry_enc', 'function_enc'
]

# Only keep columns that exist in df_fe
STRUCTURED_FEATURES = [f for f in STRUCTURED_FEATURES if f in df_fe.columns]

print(f'Structured features ({len(STRUCTURED_FEATURES)}):')
print(STRUCTURED_FEATURES)

print(f'\nText feature: text (combined corpus)')
print(f'Target: fraudulent')
```

## 12. Train / Test Split

```python
from sklearn.model_selection import train_test_split

X_text = df_fe['text']
X_struct = df_fe[STRUCTURED_FEATURES]
y = df_fe['fraudulent']

(
    X_text_train, X_text_test,
    X_struct_train, X_struct_test,
    y_train, y_test
) = train_test_split(
    X_text, X_struct, y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

print(f'Train size : {len(y_train):,}')
print(f'Test size  : {len(y_test):,}')
print(f'\nTrain class balance:')
print(y_train.value_counts())
print(f'\nTest class balance:')
print(y_test.value_counts())
```

## 13. TF-IDF Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack
import joblib

tfidf = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1, 2),       # unigrams + bigrams
    stop_words='english',
    sublinear_tf=True         # apply log normalization
)

X_tfidf_train = tfidf.fit_transform(X_text_train)
X_tfidf_test  = tfidf.transform(X_text_test)

# Combine TF-IDF + structured features
from scipy.sparse import csr_matrix

X_train_final = hstack([X_tfidf_train, csr_matrix(X_struct_train.values)])
X_test_final  = hstack([X_tfidf_test,  csr_matrix(X_struct_test.values)])

print(f'Final training matrix shape : {X_train_final.shape}')
print(f'Final test matrix shape     : {X_test_final.shape}')

# Save artefacts for modelling notebook
import os
os.makedirs('../model', exist_ok=True)

joblib.dump(tfidf,           '../model/tfidf_vectorizer.pkl')
joblib.dump(STRUCTURED_FEATURES, '../model/structured_features.pkl')
joblib.dump((X_train_final, X_test_final, y_train, y_test), '../model/train_test_data.pkl')

print('\nSaved: tfidf_vectorizer.pkl, structured_features.pkl, train_test_data.pkl')
```

## EDA Summary

| Finding | Implication |
|---|---|
| ~4.8% fake postings | Heavy imbalance → SMOTE needed, use F1/AUC not accuracy |
| Fake postings have more null fields | Missingness flags are useful features |
| Fake descriptions are shorter on average | `desc_len` is a useful numeric feature |
| Red flag keywords appear more in fake posts | `red_flag_count` adds interpretability to the Streamlit app |
| `has_company_logo=0` correlates with fraud | Binary flags are low-noise signals |
